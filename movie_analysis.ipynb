```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib
import matplotlib.pyplot as plt

# %matplotlib inline purpose is to display graphs below the cell instead of in new window
```

```python
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
# searching for missing values percentage using functions
for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print(f"{col} , {percent_missing}")
```

```python
# alternate way to find missing values
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['budget'] = df['budget'].astype(int)
df['votes'] = df['votes'].astype(int)
df['gross'] = df['gross'].astype(int)
```

```python
df
```

```python
df['year_updated'] = df['released'].str.split(',').str.get(1).str.split(' ').str.get(1)
```

```python
df
```

```python
df.sort_values(by = 'gross', inplace = False, ascending= False)
```

```python
# dropping duplicates of company names to sort the company names easily
df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
df.drop_duplicates()
```

```python
correlation = df['budget'].corr(df['gross'])
print(f"Correlation between Budget and Gross Earnings: {correlation:.2f}")
```

```python
# checking for correlation
plt.scatter(data = df, x = 'budget', y = 'gross')
plt.title("Budget vs Gross Earnings")
plt.xlabel("Movie Budget")
plt.ylabel("Gross Earnings")
plt.show()

# Since the scatter plot barely shows the correlation making us impossible to draw conclusion from 
# we will go with regression plot
```

```python
sns.regplot(x=df['budget'], y=df['gross'], scatter_kws={"color": "red"}, line_kws={"color": "blue"})

# draws a line to show relation
```

```python
df.corr(numeric_only=True, method = 'pearson')
# this avoids categorical data where there are strings only
# pearson was chosen due to linear relation (default correlation) meaning you dont have to specify method if you want to use pearson
```

```python
correlation_matrix = df.corr(numeric_only=True, method = 'pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
# Now checking for gross earnings relation to company names
df.head()
```

```python
# converting objects/strings to numeric features inside every single column that arent numeric including company names
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
# this converts object type columns to category then convert it to numeric
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True, method = 'pearson')
# Create the heatmap with improved settings
sns.heatmap(
    correlation_matrix, 
    annot=True, 
    cmap='coolwarm', 
    fmt=".2f", 
    linewidths=0.5,
    annot_kws={"size": 6}  # Adjust text size inside the heatmap
)
plt.title("Correlation Matrix for Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
# Set figure size for better readability
plt.figure(figsize=(18, 20))


plt.show()
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```
